```python
import pandas as pd
import json
```

```python
#opean the dataset 
with open('./yelp_dataset/business.json') as f:
    df = pd.DataFrame(json.loads(line) for line in f)
```

```python
## clean dataset

#  1.filter to Las Vegas dataset and get useful columns  
data = df[df.city == 'Las Vegas']
data = data[['business_id','categories','city','latitude','longitude','name','postal_code','review_count','stars','state']]

#print(data.info())
data.head()
## 2.zipcode , get map true zip code from Las Vegas 

LV_zipcode = [89002, 89004, 89005, 89007, 89011, 89012, 89014, 89015, 89018, 89019, 89021, 89025, 89026, 89027, 89029, 89030, 89031, 89032, 89039, 89040, 89044, 89046, 89052, 89054, 89074, 89081, 89084, 89085, 89086, 89101, 89102, 89103, 89104, 89106, 89107, 89108, 89109, 89110, 89113, 89115, 89117, 89118, 89119, 89120, 89121, 89122, 89123, 89124, 89128, 89129, 89130, 89131, 89134, 89135, 89138, 89139, 89141, 89142, 89143, 89144, 89145, 89146, 89147, 89148, 89149, 89156, 89161, 89166, 89169, 89178, 89179, 89183, 89191]
## useless zip code
LV_zipcode.remove(89026)
LV_zipcode.remove(89191)

data = data[data.postal_code.isin([''])== False]
data.postal_code = data.postal_code.astype('int64')
data[data.postal_code.isin(LV_zipcode)]
#print(data.info())

## 3.missing data , drop missing dataset, as we can see categories has missing data
data = data.dropna()
print(data.info())
print(data.describe())
data.head()

```

```python
categories_all = []

data.categories.tolist()[0].split(',')

for i in data.categories.tolist():
    try:
        categories_all.extend(i.split(','))
    except:
        pass
    
categories_uniuqe = list(set(categories_all))

count_categories = {}
for name in categories_all:
     count_categories[name] = count_categories.get(name, 0) + 1
        
sorted(count_categories.items(), key=lambda kv: kv[1],reverse=True)
```
